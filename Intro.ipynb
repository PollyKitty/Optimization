## Виды матриц

Исторически сложилось, что в зависимости от того, какие матрицы (наполнение матриц) - такие и существуют полезные свойства при дальнейшем анализе и решении задач. В основном сейчас понадобятся следующие матрицы:

### Симметричные матрицы

**Транспонирование:** $A^{\top}$, данная операция меняет порядок индексирования в матрице $A_{i,j} \rightarrow A_{j,i}$

Комплексное сопряжение: $A^{*}$, для комплексных матриц меняет порядок индексирования + меняет комплексные числа на их комплексное споряжение -   $A_{i,j} \rightarrow A_{j,i}, \forall z = x \mp iy \rightarrow x \pm iy$ 

Таким образом, если  $A$ - вещественная матрица, то $A^{*} = A^{\top}$

$A = A^{\top}$ - **симметричные матрицы**

#### Билинейная и квадратичная форма

Чтобы описать практическое применение симметричных матриц и их свойства стоит сначала немного погрузиться в терминологию, которая идет бок о бок с понятием симметричности. Это будет легкое введение, так как в основе данного подраздела лежат основы алгебры, на что не хочется делать сильный акцент. 

Существуют способы задания объектов в пространстве, которые обобщенно называют формами.

*Линейная форма*

Линейная форма это какое-то **линейное отображение** из вектора $f : \mathbb{R} \rightarrow \mathbb{R}$. В общем случае, линейная форма - это вектор $a = \begin{pmatrix} a_1, & \cdots, & a_n \end{pmatrix}$ для вектора $x = \begin{pmatrix} x_1 \\ \vdots \\ x_n \end{pmatrix}$, вектор $a$ является линейной формой, а $f_a(x) = ax = \begin{pmatrix} a_1, & \cdots, & a_n \end{pmatrix}\begin{pmatrix} x_1 \\ \vdots \\ x_n \end{pmatrix} = a_1x_1 + \cdots + a_nx_n$ является линейным отображением, примером линейной формы может быть среднее $\textbf{mean}(x) = [\frac{1}{n}, \cdots , \frac{1}{n}]x$

*Билинейная форма*

Билинейная форма это какое-то **линейное отображение** из векторов $f : \mathbb{R} \times \mathbb{R} \rightarrow \mathbb{R}$. Не стоит путать наличие различных векторов записи, например $f(x,y) = x + y$ - линейная форма, $f(x, y) = xy$ - билинейная форма.

Пример.

$$
A = \begin{pmatrix}
1 & 2 \\
3 & 4
\end{pmatrix} \\
f\begin{pmatrix}\begin{pmatrix}x_1 \\ y_1\end{pmatrix},  \begin{pmatrix}x_2 \\ y_2\end{pmatrix}\end{pmatrix} = \begin{pmatrix}x_1 & y_1\end{pmatrix} \begin{pmatrix}
1 & 2 \\
3 & 4
\end{pmatrix} \begin{pmatrix}x_2 \\ y_2\end{pmatrix} = x_1x_2 +2x_1y_2 + 3y_1x_2 + 4y_1y_2
$$

$B(x,y) = \sum_{i,j=1}^n x_iA_{ij}x_j = x^{\top}Ay$ - запись билинейной формы в матричной нотации.

Пример восстановления билинейной формы по заданному функционалу.

$$
f\begin{pmatrix}\begin{pmatrix}x_1 \\ y_1\end{pmatrix},  \begin{pmatrix}x_2 \\ y_2\end{pmatrix}\end{pmatrix}  = 2x_1x_2 + 3x_1y_2 + x_2y_1 \\
A = \begin{pmatrix}
2 & 3 \\
1 & 0
\end{pmatrix}
$$

*Базис*

Эта часть сделана для напоминания, ее вставка преследует идею заставить читателя вспомнить про некоторые базовые формулы из линейной алгебры, которые при недостаточно понимании можно освежить из базового курса.

Базис - это такой набор векторов в пространстве, что любой вектор из данного пространства представим ввиде линейной комбинации (линейного отображения) векторов из базиса. Классический базис пространства, если нигде не упоминается базис это единичная матрица, для пространства $\mathbb{R}^2$ это вектора $e_1 = \begin{pmatrix}1 \\ 0\end{pmatrix},  e_2 = \begin{pmatrix}0 \\ 1\end{pmatrix}$, действительно через сумму этих векторов с какими-то коэффициентами можно представить абсолютно любой вектора из $\mathbb{R}^2$ пространства.

Теперь предположим, что существует два базиса $\mathbb{C} = \{c_1, \cdots, c_n\}, \mathbb{B} = \{b_1, \cdots, b_n\}$, которые описываются своими базисными векторами, то мы можем написать один базис в терминах другого $c_i = \sum_{j=1}^n m_{j,i}b_j$, или 

$\mathbb{C} = M\mathbb{B}$, 

где матрица 

$$
M = \begin{pmatrix}
m_{1,1} & \cdots & m_{1,n} \\
\vdots & & \vdots \\
m_{n,1} & \cdots & m_{n,n}
\end{pmatrix}
$$

называется матрицей перехода от базиса $\mathbb{B}$ к базису $\mathbb{C}$. И для любого вектора $v$ будет верно

$v_{\mathbb{B}} = Mv_{\mathbb{C}}$

Действительно, если $\mathbb{B}$ и $\mathbb{C}$ - базисы то абсолютно любой вектор может быть представлен как какая-то линейная комбинация базисных векторов, тогда вектор $\overrightarrow{v}$ может выражаться в терминах первого и второго базиса $\overrightarrow{v} = \mathbb{B}v_{\mathbb{B}} = \mathbb{C}v_{\mathbb{C}}$, которое можно записать $v_{\mathbb{C}}\mathbb{C} = v_{\mathbb{B}}\mathbb{B} \rightarrow v_{\mathbb{B}} = Mv_{\mathbb{C}}$.

В свою очередь линейное преобразование выражается следующим образом: $[T]_{\mathbb{C}} = M^{-1}[T]_\mathbb{B}M$

*Теорема о смене базиса. Если $\mathbf{f}$ билинейная форма, то существует следующее соотношение для базисов $\mathbb{B}, \mathbb{C}$, где матрица $M$ - матрица перехода от базиса $\mathbb{B}$ к базису $\mathbb{C}$*

$\mathbf{[f]_{\mathbb{C}} = M^{\top}[f]_{\mathbb{B}}M}$

*Определение. Билинейная форма называется симметричной, если удовлетворяет условию $\mathbf{f(u,w) = f(w,u)}$, для заданной симметричной билинейной есть ассоциированная ей квадратичная форма $\mathbf{q(v) = f(v,v)}$*

*Теорема о поляризации. Для любой квадратичной формы существует уникальная симметричная билинейная форма ассоциированная ей.*

### Симметричные матрицы. Продолжение

Для каждой симметричной матрицы существует квадратичная форма $x^{\top}Ax$, квадратичные формы появляются во многих задачах оптимизации, а также в выводе второй производной, что в дальнейшем будет важно $F(x + u) = F(x) + F'(x)u +\frac{1}{2}u^{\top}H(x)u + O(||u||^3)$, где в стационарной точке $F'(x) = 0$, $H$ - положительно-опредленная (стационарная точка - точка минимума) или отрицательно-определенная (стационарная точка - точка максимума) матрица, если $H$ - неопределена, то есть имеет и отрицательные и положительные **собственные значения**, то в таком случае точка называется седловой.

#### Закон инерции Сильвестра

Немного про существующин соотношения между матрицами, если для матриц $A, B$ найдется невырожденая матрица $P$, такая что $B = P^{-1}AP$, то эти матрицы называются **подобными**, если соотношение такое $B = P^{T}AP$, то такие матрицы называются **конгурентными**.

Как мы увидели ранее любой симметричной билинейной форме соответсвует квадратичная форма. Одной из самых распространенных билинейных форм является матрица Грамма (о которой позже), данная матрица задает операцию скалярного произведения. Таким образом, в случае матрицы Грамма, матрица $B = P^{T}AP$ является скалярным произведением (билинейной формой) в другом базисе. Иногда необходимо быстро находить подобие матриц на таком уровне (схожесть билинейных форм с точности до базиса), но не всегда удобно в лоб подбирать такую матрицу $P$, в таком случае на помощь приходит закон инерции Сильвестра.

Что такое инерция матрицы? Зависит от источника, как правило инерция $In(a)$ это тройка - {кол-во позитивных собственных чисел, негативных, нулевых}. Закон инерции Сильвестра утверждает, что если матрицы являются конгурентными, то их инерции совпадают. А обобщеннее ...

**Теорема. Закон инерции Сильвестра. Если матрица $\mathbf{A}$ симметрична и матрица $\mathbf{B}$ невырождена, то $\mathbf{In(A) = In(B^{\top}AB)}$**

Также данный закон формулируют через понятие о **каноничной форме матрицы**, для квадратинчой формы существует всего одна вещественная каноничная форма.

**Каноничная форма (Жарданова форма)** - это особый способ блочного представления матрицы, который используется для оптимизации численных процессов (что-то решить по блокам), либо в разложениях.

Если **симметричная матрица положительно определена** $\forall x \neq 0: x^{\top}Ax > 0$ то она и называется соотвественно (Symmetric Positive Defined).

И у нее есть некоторые полезные свойства, например:

1. У матрицы $T = \frac{A + A^{\top}}{2}$ только положительные собственные значения
2. $A$ - PSD, $A^{-1}$ - PSD
3. Любой минор PSD - PSD
4. Любое дополнение Шура - PSD
5. Для любого минора $M$ матрицы $A$, $\mathbb{k_2}(M) \leq \mathbb{k_2}(A)$, где $\mathbb{k_2}$ - **число обусловленности**

$\sigma_1(A) = \underset{||x||_2 = 1}{\max} x^{\top}Ax, \sigma_{\min} = \underset{||x||_2 = 1}{\min} x^{\top}Ax$, где $\sigma_1, \sigma_{\min}$ - наибольшее и наименьшее **сингулярные значения матрицы** (НУЖЕН БРИФ ПРО СОБСТВЕННЫЕ ЧИСЛА ИХ СВЯЗЬ С СИНГУЛЯРНЫМИ) допустим $M = A_{11}$, тогда $\sigma_1(M) = \underset{||x||_2 = 1}{\max} x^{\top}Mx = \underset{||x||_2 = 1}{\max} \begin{bmatrix}x \\ 0 \end{bmatrix}^{\top}A\begin{bmatrix}x \\ 0 \end{bmatrix} \leq \underset{||z||_2 = 1}{\max} z^{\top}Az = \sigma_1(A)$, тоже верно и для $\sigma_{\min}$ 

$\mathbb{k_2}(M) = \frac{\sigma_1(M)}{\sigma_{\min}(M)} \leq \frac{\sigma_1(A)}{\sigma_{\min}(A)} = \mathbb{k_2}(A)$
